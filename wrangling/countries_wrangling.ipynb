```python
# import packages
import pandas as pd
import geopandas as gpd
import numpy as np
```

```python
# geojson countains geo and descriptive data for all countries
all_countries = gpd.read_file('source data/all_countries.geojson')  
# csv contains centroids of countries, will be merged with all_countries
country_centroids = pd.read_csv("source data/country_centroids.csv")

# change country names to match country centroids
all_countries['GEM_REGION'] = all_countries['GEM_REGION'].str.replace('_', ' ')
all_countries['COUNTRY'] = all_countries['COUNTRY'].str.replace('_', ' ')
all_countries['COUNTRY'] = all_countries['COUNTRY'].str.replace('Czechia', 'Czech Republic')
all_countries['COUNTRY'] = all_countries['COUNTRY'].str.replace('Macao', 'Macau')
all_countries['COUNTRY'] = all_countries['COUNTRY'].str.replace('North Macedonia', 'Macedonia')

# rename columns
all_countries = all_countries.rename(columns = {
    'GEM_REGION': 'REGION',
    'BUILDINGS_TOTAL': 'BUILDINGS TOTAL', 
    })

# select columns
all_countries = all_countries[['ISO3', 'COUNTRY', 'CONTINENT', 'REGION', 'BUILDINGS TOTAL', 'geometry']]
# merge with country centroids
all_countries = all_countries.merge(country_centroids, how = 'left', left_on = 'COUNTRY', right_on = 'country').drop(["country"], 1)
```

```python
## data from World Bank (2021)
# csv contains pop_density of countries
pop_totl = pd.read_csv("source data/WB_POP_TOTL.csv")
# csv contains pop_density of countries
pop_density = pd.read_csv("source data/WB_POP_DNST.csv")
# csv contains pop_density of countries
gdp_usd = pd.read_csv("source data/WB_GDP_USD.csv")
# csv contains pop_density of countries
gdp_per_cap = pd.read_csv("source data/WB_GDP_PCAP.csv")
# csv contains pop_density of countries
gov_effect = pd.read_csv("source data/WB_GOV_EFCT_RNK.csv")
```

```python
# rename 2021 column to population density (log)
pop_totl['POPULATION'] = pop_totl['2021'].round(2)
# rename 2021 column to population density (log)
pop_density['POP. DENSITY (LOG)'] = np.log(pop_density['2021'].round(2))
# rename 2021 column to population density (log)
gdp_usd['GDP (USD)'] = gdp_usd['2021'].round(2)
# rename 2021 column to population density (log)
gdp_per_cap['GDP PER CAPITA'] = gdp_per_cap['2021'].round(2)
# rename 2021 column to population density (log)
gov_effect['GOV. EFFECTIVENESS'] = gov_effect['2021'].round(2)
```

```python
# merge with pop_density
all_countries = all_countries.merge(pop_totl[['Country Code', 'POPULATION']], how = 'left', left_on = 'ISO3', right_on = 'Country Code').drop(["Country Code"], 1)
all_countries = all_countries.merge(pop_density[['Country Code', 'POP. DENSITY (LOG)']], how = 'left', left_on = 'ISO3', right_on = 'Country Code').drop(["Country Code"], 1)
all_countries = all_countries.merge(gdp_usd[['Country Code', 'GDP (USD)']], how = 'left', left_on = 'ISO3', right_on = 'Country Code').drop(["Country Code"], 1)
all_countries = all_countries.merge(gdp_per_cap[['Country Code', 'GDP PER CAPITA']], how = 'left', left_on = 'ISO3', right_on = 'Country Code').drop(["Country Code"], 1)
all_countries = all_countries.merge(gov_effect[['Country Code', 'GOV. EFFECTIVENESS']], how = 'left', left_on = 'ISO3', right_on = 'Country Code').drop(["Country Code"], 1)

# select columns 
all_countries = all_countries[['ISO3', 'COUNTRY', 'CONTINENT', 'REGION', 'POPULATION', 'POP. DENSITY (LOG)', 
                               'GDP (USD)',  'GDP PER CAPITA', 'GOV. EFFECTIVENESS', 'BUILDINGS TOTAL', 'center_lat', 'center_lon', 'geometry']]
```

```python
# add missing data for Venezuela (source: World Bank)
all_countries.loc[all_countries['ISO3'] == 'VEN', 'GDP (USD)'] = 92000000000
all_countries.loc[all_countries['ISO3'] == 'VEN', 'GDP PER CAPITA'] = 3470
all_countries.loc[all_countries['ISO3'] == 'VEN']
```

```python
# method sets the zoom level for a given country    
# method called createCenters method
def setZoom(poly):
    x1, y1, x2, y2 = poly.bounds
    max_bound = max(abs(x1 - x2), abs(y1 - y2)) * 80
    zoom = 11.5 - np.log(max_bound)
    return zoom

for i in range(len(all_countries)):
    all_countries['zoom'] = all_countries['geometry'].apply(setZoom, 1)
```
